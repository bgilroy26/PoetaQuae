```python
virgilurls=[]
for i in range(1,9):
    virgilurls.append('https://www.sacred-texts.com/cla/virgil/aen/aenl0{}.htm'.format(i))
virgilurls.append('https://www.sacred-texts.com/cla/virgil/aen/aenl10.htm')
virgilurls.append('https://www.sacred-texts.com/cla/virgil/aen/aenl11.htm')
virgilurls.append('https://www.sacred-texts.com/cla/virgil/aen/aenl12.htm')
for i in range(1,9):
    virgilurls.append('https://www.sacred-texts.com/cla/virgil/ecl/ecll0{}.htm'.format(i))
virgilurls.append('https://www.sacred-texts.com/cla/virgil/ecl/ecll10.htm')
for i in range(1,4):
    virgilurls.append('https://www.sacred-texts.com/cla/virgil/geo/geol0{}.htm'.format(i))
```

```python
virgil=[]
for url in virgilurls:
    response=requests.get(url)
    soup=BeautifulSoup(response.content,'html.parser')
    data=soup.find_all('p')
    for d in data:
        virgil.append(d.text)
```
