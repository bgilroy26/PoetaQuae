```python
catul=[]
for c in catullus:
#     ad=c.strip('\n</catullus_translation>').strip('<catullus_translation>\r\n').strip('<br/>').strip('* * * * * * * ')
#     ad=c.strip('<catullus_translation>\r\n')
    ad=c.replace('<br/>\r\n',' ')
    ad=ad.replace('<catullus_translation>','')
    ad=ad.replace('\r\n','')
    ad=ad.replace('<br/>\n','')
    ad=ad.replace('</catullus_translation>','')
    ad=ad.replace('\x97','a')
#     ad=ad.replace(r'<br/>\n',' ')
#     ad=ad.replace('\x97','a ')
    if ad!=' ':
        if ad!='':
            if ad!='  ':
                if len(ad)>1:
                    catul.append(ad)
```

```python
catul
```

```python
catulluspoems=[]
catullusp=[]
for c in catul:
    cat=c.split('?')
    catullusp.append(cat)
flatcatullus=[]
for sublist in catullusp:
    for item in sublist:
        flatcatullus.append(item)
for c in flatcatullus:
    cat=c.split('!')
    catulluspoems.append(cat)
```
